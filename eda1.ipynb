```python
missing = df.isnull().sum().to_frame("missing_count")
missing["missing_ratio"] = (missing["missing_count"] / len(df)).round(4)

missing.sort_values("missing_ratio", ascending=False)
```

```python
import matplotlib.pyplot as plt

missing_ratio = missing["missing_ratio"].sort_values(ascending=False).head(20)

plt.figure(figsize=(10,5))
missing_ratio.plot(kind="bar")
plt.title("Top 20 Missing Value Ratios")
plt.ylabel("Ratio")
plt.show()
```
